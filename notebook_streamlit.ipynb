```python
def convertir_fecha_a_espanol(df, columna_fecha):
    # Convertir la columna 'columna_fecha' a tipo datetime
    df[columna_fecha] = pd.to_datetime(df[columna_fecha])

    # Crear una nueva columna formateada como 'Mes Año' en inglés
    df[columna_fecha] = df[columna_fecha].dt.strftime('%B %Y')

    # Diccionario de traducción de los meses de inglés a español
    meses_traduccion = {
        'January': 'Enero', 'February': 'Febrero', 'March': 'Marzo',
        'April': 'Abril', 'May': 'Mayo', 'June': 'Junio',
        'July': 'Julio', 'August': 'Agosto', 'September': 'Septiembre',
        'October': 'Octubre', 'November': 'Noviembre', 'December': 'Diciembre'
    }

    # Función para traducir los nombres de los meses
    def traducir_mes(fecha):
        for eng, esp in meses_traduccion.items():
            fecha = fecha.replace(eng, esp)
        return fecha

    # Aplicar la función de traducción a la columna
    df[columna_fecha] = df[columna_fecha].apply(traducir_mes)

    return df
```

```python
# Ejemplo de uso
data = {'Fecha': ['2024-04-01 00:00:00', '2024-05-01 00:00:00', '2024-06-01 00:00:00']}
df = pd.DataFrame(data)

df = convertir_fecha_a_espanol(df, 'Fecha')

print(df)
```
